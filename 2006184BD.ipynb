```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn import datasets
import json

tmdbCredits = pd.read_csv("tmdb_5000_credits.csv")
tmdbMovies = pd.read_csv("tmdb_5000_movies.csv")

tmdbCredits.head()

print(tmdbCredits['crew'])
#print(tmdbMovies.describe())
#print(type(boston))
#plt.hist(y, bins=30, color='b', density=True, stacked=True)
```

```python
tmdbMovies.head()
```

```python
tmdbCredits.columns = ['id','title','cast','crew']
tmdb = tmdbMovies.merge(tmdbCredits, on='id')
tmdb = tmdb.drop(['title_y','title_x','homepage','tagline', 
                  'id', 'budget', 'revenue', 'status','spoken_languages', 
                  'original_language', 'overview', 'production_companies', 
                  'production_countries', 'runtime'], axis=1)
```

```python
tmdb.dtypes
```

```python
tmdb.count()
```

```python
dict_columns = ['genres','keywords','cast','crew']
for column in dict_columns:   
    tmdb[column] = tmdb[column].apply(json.loads)
```

```python
def extract_genres(genre_info):
    genres = []
    for genre in genre_info:
        genres.append(genre['name'])
    if len(genres) == 0:
        return np.nan
    else:
        return genres
    

tmdb['genres'] = tmdb['genres'].apply(extract_genres)
```

```python
def extract_actor(cast_info, order):
    for cast_member in cast_info:
        if cast_member['order'] == order:
            return cast_member['name']
    return ''

tmdb['actor_1'] = tmdb['cast'].apply(extract_actor, order=0)
tmdb['actor_2'] = tmdb['cast'].apply(extract_actor, order=1)
tmdb['actor_3'] = tmdb['cast'].apply(extract_actor, order=2)

tmdb = tmdb.drop(['cast'], axis=1)
```

```python
def extract_cast(crew_info, job):
    cast = []
    for crew_member in crew_info:
        if crew_member['job'] == job:
            cast.append(crew_member['name'])
    if len(cast) == 0:
        return np.nan
    else:
        return cast

tmdb['director'] = tmdb['crew'].apply(extract_cast, job='Director')
tmdb['producer'] = tmdb['crew'].apply(extract_cast, job='Producer')
tmdb['editor'] = tmdb['crew'].apply(extract_cast, job='Editor')

tmdb = tmdb.drop(['crew'], axis=1)
```
